```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
go_headers={ 
            'pragma': 'no-cache',
            'cache-control': 'no-cache',
            'upgrade-insecure-requests': '1',
            'user-agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Mobile Safari/537.36',
            'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
            'sec-fetch-site': 'none',
            'sec-fetch-mode': 'navigate',
            'sec-fetch-user': '?1',
            'sec-fetch-dest': 'document',
            'accept-language': 'ru-RU,ru;q=0.9'}
```

```python
def pars_page(i_numberPage):
    
    df = pd.DataFrame()
    with requests.Session() as lo_s:

        lo_r = lo_s.get("https://auto.drom.ru/all/page"+str(i_numberPage)+"/?minyear=2022", headers = go_headers)
        if (lo_r.status_code == 200):
    #         with open('xz.html', 'w') as file:
    #             file.write(lo_r.text)
            #lo_r.encoding = 'utf-8'
            soup = BeautifulSoup(lo_r.text, "html.parser")
            allcar = soup.findAll('a', class_='css-1dk6v6x ewrty960')
            for data in allcar:
                try:
                    dic = {}
                    #print(data.find('div', class_='css-17lk78h e3f4v4l2').text)
                    dic["Авто"] = data.find('div', class_='css-17lk78h e3f4v4l2').text
                    dic["Цена"] = data.find('span', class_='css-107ouu3 e162wx9x0').text
                    dic["Состояние"] = data.find('div', class_='css-citfcw ejipaoe0').text
                    lst_dv = data.findAll('span', class_='css-159lyxl e162wx9x0')
                    dic["Двигатель"] = lst_dv[0].text
                    dic["Топливо"] = lst_dv[1].text
                    dic["КПП"] = lst_dv[2].text
                    dic["Привод"] = lst_dv[3].text
                    # print(dic)
                    df = df.append(dic, ignore_index=True)
                except Exception:
                    print('Это что ещё такое?')
            df.head()

            #print(len(allcar))

        else:
            print("Error", lo_r.status_code)
    return df
```

```python
pars_page(5)
```

```python
df_full = pd.DataFrame()
for i in range(2, 10):
    print(i)
    a = pars_page(i)
    df_full = df_full.append(a)
df_full.to_excel('Drom.xlsx')
```
